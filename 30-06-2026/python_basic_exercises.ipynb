```python

#Q8.  Counting Spam Predictions


def count_spam(predictions):
    count = 0

    for prediction in predictions:
        if prediction.strip().lower() == "spam":
            count += 1

    return count

user_input = input("Enter the predictions separated by commas: ")

# Convert input into a list
predictions = user_input.split(",")

# Count spam predictions
spam_count = count_spam(predictions)

print("\nPrediction List:", predictions)
print("Total Spam Predictions:", spam_count)

print("\nThank you for using the Spam Prediction Counter!")
```

```python
#Q16. Encapsulation — Bank Account
class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance  # private variable

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"{amount} deposited successfully.")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"{amount} withdrawn successfully.")
        else:
            print("Insufficient balance or invalid amount.")

    def check_balance(self):
        print("Current Balance:", self.__balance)


# Example usage
account = BankAccount(1000)

account.deposit(500)
account.withdraw(300)
account.check_balance()
```

```python
#Q19. Abstraction — Model

from abc import ABC, abstractmethod

# Abstract class
class Model(ABC):

    @abstractmethod
    def train(self):
        pass


# Child class 1
class SpamDetector(Model):
    def train(self):
        print("Training Spam Detector model...")


# Child class 2
class ImageClassifier(Model):
    def train(self):
        print("Training Image Classifier model...")


# Creating objects
model1 = SpamDetector()
model2 = ImageClassifier()

# Calling train() method
model1.train()
model2.train()
```

```python
# Q23. Comparing AI Models with __eq__

class AIModel:
    def __init__(self, model_name, accuracy):
        self.model_name = model_name
        self.accuracy = accuracy

    def __eq__(self, other):
        return self.accuracy == other.accuracy


# Example usage
model1 = AIModel("Spam Detector", 95.5)
model2 = AIModel("Image Classifier", 95.5)
model3 = AIModel("Face Recognition", 97.2)

print(model1 == model2)  # True (same accuracy)
print(model1 == model3)  # False (different accuracy)
```

```python
# Q30. AI Spam Detection System
import logging
import random

# Setup logging
logging.basicConfig(
    filename="ai_system.log",
    level=logging.INFO,
    format="%(asctime)s Prediction : %(message)s"
)

class SpamDetector:
    def __init__(self, model_name, version):
        self.model_name = model_name
        self.version = version

    def predict(self, message):
        # Exception handling
        if message == "":
            raise Exception("Message cannot be empty")

        # Simple spam detection logic
        spam_words = ["win", "free", "prize", "money", "click"]

        if any(word in message.lower() for word in spam_words):
            result = "Spam"
            confidence = random.randint(85, 98)
        else:
            result = "Not Spam"
            confidence = random.randint(80, 97)

        # Save prediction to file
        file = open("prediction.txt", "a")
        file.write(f"{message} -> {result} ({confidence}%)\n")
        file.close()

        # Save to log file
        logging.info(f"{result} Confidence : {confidence}%")

        return result, confidence


# ---------------- Example Usage ----------------

detector = SpamDetector("SpamAI", "1.0")

try:
    print(detector.predict("You won a free prize"))
    print(detector.predict("Hello, how are you?"))
    print(detector.predict(""))  # This will raise exception

except Exception as e:
    print("Error:", e)
```

```python
#Q35. OOP + File Handling A Student class was created, but every time the program ends, all student information is lost. Which Python concept would you use to make the data available even after the program closes? Explain your approach and write the code.

#I would use File Handling to store the student information in a file such as students.txt.
#  Since files are stored on disk, the data remains available even after the program closes.
#  I use append mode ('a') so that new student records are added without deleting the existing ones."


#CODE:

class Student:
    def __init__(self, name, age, course):
        self.name = name
        self.age = age
        self.course = course

    def save_data(self):
        file = open("students.txt", "a")
        file.write(f"{self.name}, {self.age}, {self.course}\n")
        file.close()


student1 = Student("Rahul", 20, "AI")
student1.save_data()

print("Data saved successfully.")


```
