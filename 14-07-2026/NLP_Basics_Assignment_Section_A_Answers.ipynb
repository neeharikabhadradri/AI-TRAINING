## Section A – Theoretical Questions

## T1. What is Natural Language Processing (NLP)? Explain in your own words how it acts as a bridge between human language and computers.

Natural Language Processing (NLP) is a branch of Artificial Intelligence (AI) that enables computers to understand, analyze, and generate human language. It combines computer science, linguistics, and machine learning to process text and speech.

NLP acts as a bridge between humans and computers by converting human language into a format that machines can understand. It helps computers identify words, understand meanings, and respond appropriately.

## T2. Why do computers need NLP to understand human language? Give one real-world example to support your answer.

Computers understand numbers and structured data, while human language is complex. NLP converts text into a machine-readable form.

Example: Google Translate uses NLP to translate text between languages.

## T3. List any four real-life applications of NLP and briefly describe what each one does.

1. Chatbots – Answer customer queries.
2. Machine Translation – Translates languages.
3. Sentiment Analysis – Detects opinions.
4. Spam Detection – Identifies spam messages.

## T4. What is a rule-based NLP system? Explain using the spam detection example.

A rule-based NLP system uses predefined rules. Example: If a message contains 'Congratulations', classify it as 'Possible Spam'; otherwise 'Not Spam'.

1. Text Collection
2. Text Cleaning
3. Lowercasing
4. Tokenization
5. Stopword Removal
6. Feature Extraction
7. Model Building and Prediction

## T6. What is tokenization? Why is it necessary?

Tokenization splits text into words or tokens. It is necessary because ML models cannot process raw text directly.

## T7. What are stopwords? Give three examples.

Stopwords are common words such as 'the', 'is', and 'and'. They are usually removed because they add little meaning.

## T8. Why convert text to lowercase?

Lowercasing treats 'Apple', 'APPLE', and 'apple' as the same word, improving consistency.
